```python
import matplotlib.pyplot as plt
import numpy as np

print("=" * 50)
print("=" * 50)

class GridPathfinderV1:
    def __init__(self):
        self.rows = 8
        self.cols = 10
        self.init_grid()
        
    def init_grid(self):
        self.grid = np.zeros((self.rows, self.cols))
        self.start = (6, 0)
        self.target = (6, 7)
        
        # Set walls
        walls = [(1, 5), (2, 5), (3, 5), (4, 5), (4, 7)]
        for r, c in walls:
            self.grid[r, c] = -1
            
        self.grid[self.start] = 1
        self.grid[self.target] = 2
    
    def visualize(self):
        fig, ax = plt.subplots(figsize=(10, 8))
        
        # Simple visualization
        colors = ['red', 'white', 'blue', 'green']
        cmap = plt.cm.colors.ListedColormap(colors)
        
        ax.imshow(self.grid, cmap=cmap, vmin=-1, vmax=2)
        
        # Add labels
        for i in range(self.rows):
            for j in range(self.cols):
                val = int(self.grid[i, j])
                if val == 1:
                    ax.text(j, i, 'S', ha='center', va='center', fontsize=14, fontweight='bold')
                elif val == 2:
                    ax.text(j, i, 'T', ha='center', va='center', fontsize=14, fontweight='bold')
                elif val == -1:
                    ax.text(j, i, 'â–ˆ', ha='center', va='center', fontsize=14)
        
        ax.set_title("Version 1: Basic Grid with Start (S) and Target (T)")
        ax.grid(True)
        plt.show()

v1 = GridPathfinderV1()
v1.visualize()
```
